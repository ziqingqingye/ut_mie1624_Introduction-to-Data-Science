## **Data Cleanning**

```python
from nltk.stem.porter import PorterStemmer
des_clean=[]
stop_words = set(stopwords.words("english"))
for i in range(0, 1100):
      #Remove punctuations
    text = re.sub('[^a-zA-Z]', ' ', df['Descriptions'][i])
    
    #Convert to lowercase
    text = text.lower()
    
    #remove tags
    text=re.sub("&lt;/?.*?&gt;"," &lt;&gt; ",text)
    
    # remove special characters and digits
    text=re.sub("(\\d|\\W)+"," ",text)
    
    ##Convert to list from string
    text = text.split()
    
    ##Stemming
    ps=PorterStemmer()
    #Lemmatisation
    lem = WordNetLemmatizer()
    text = [lem.lemmatize(word) for word in text if not word in  
            stop_words] 
    text = " ".join(text)
    des_clean.append(text)
df['Descriptions_clean']=des_clean
```

# **Hierarchical clustering implementation**

Hierarchical clustering is achieved by using Euclidean distance between single observations in data set and linkage criterion. First, calculate the Euclidean distance between each skills. Then using the dendrogram to plot the clustering result with Ward linkage. In order to design 8-12 courses, i set the threshold line at index 13 which gives 8 intersection points with the dendrogram, which indicates 8 clusters.

```python
df_hi = df_new.iloc[:,5:]
df_hi.T.index.tolist()
```

```python
#dist_sin = linkage(df_hi.T, method='ward', metric='euclidean')
plt.figure(figsize=(15,6))
import scipy.cluster.hierarchy as sch
# rotates the x axis labels
# font size for the x axis labels
#Link the clusters
# plot the dendrogram
dendrogram=sch.dendrogram(sch.linkage(df_hi.T, method='ward', metric='euclidean'), labels = df_hi.T.index.tolist(), leaf_rotation=90,color_threshold=13) #Set thereshold at 13)
#keep 8 clusters
plt.axhline(y=13, c='grey', lw=1, linestyle='dashed')
plt.xlabel('skills')
plt.ylabel('Distance')
plt.suptitle("DENDROGRAM SINGLE METHOD",fontsize=18)
plt.show()
```

# **K-means or DBSCAN clustering implementation**

Another method that can help us with unsupervised clustering is k-means clustering. We would assign each observation to the cluster with the closest mean and recompute the means for the observations assigned to each cluster, repeating the steps until they converge. First, to decide how many groups we should use in the algorithm K means. I use the elbow method to find a compromise between inertia and number of groups. We may find that 8 or 9 groups might be appropriate.Then we apply the K-Means algorithm with 9 groups and plot the scatterplot to visualize the results.

```python
from sklearn.cluster import KMeans
data = df_hi
distortions = []



for i in range(1,15):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data)
    distortions.append(kmeans.inertia_)

plt.plot(range(1,15), distortions, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters=8)
kmeans = kmeans.fit(data)
```

```python
data['cluster'] = kmeans.fit_predict(data)
data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
pca_num_components = 2

reduced_data = PCA(n_components=pca_num_components).fit_transform(data)
results = pd.DataFrame(reduced_data,columns=['pca1','pca2'])

sns.scatterplot(x="pca1", y="pca2", hue=data['cluster'], data=results)
plt.title('K-means Clustering with 2 dimensions')
plt.show()
```
